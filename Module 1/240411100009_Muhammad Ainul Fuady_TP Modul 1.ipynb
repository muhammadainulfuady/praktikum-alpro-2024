# 2. Badan Sensus Penduduk memproyeksikan populasi suatu negara berdasarkan asumsi berikut:
- Satu kelahiran setiap 7 detik
- Satu kematian setiap 13 detik
- Satu migrasi/pindah setiap 45 detik
# Tuliskan suatu program untuk menampilkan populasi pada setiap lima tahun berikutnya. Anggaplah polulasi saat ini adalah 250.000.000 dan setiap tahun terdapat 365 hari. Contoh:
Screenshot 2024-09-27 144138.png

```python
# NOMER 2
populasiawal = 250000000
Detik = 365 * 24 * 60 * 60
kelahiran = Detik//7
kematian = Detik//13
imigrasi = Detik//45
perubahan = kelahiran - kematian - imigrasi
print(kelahiran)
print(kematian)
print(imigrasi)
print(perubahan)
populasiakhir = (populasiawal + (perubahan * 5))
print (f"Populasi dalam 5 tahun depan= {populasiakhir}")
```

# 3. Suatu program menampilkan konversi suhu Celcius ke Reamur, Fahrenheit, Rankine dan Kelvin dimana inputan Celcius merupakan inputan dari user seperti gambar 3a dan 3b.
Screenshot 2024-09-27 191335.png

```python
# NOMER 3
celcius = float(input("Masukkan Suhu Celcius : "))
celtofar = (9/5 * celcius) + 32
celtokelf = celcius + 273.15
celtorea = 4/5 * celcius
celtorank = (celcius + 273.15) * 9/5
print (f'Konverensi Suhu Celcius Ke Farenheit  : {celtofar}')
print (f'Konverensi Suhu Celcius Ke Kelvin : {celtokelf}')
print (f'Konverensi Suhu Celcius Ke Reamor : {celtorea}')
print (f'Konverensi Suhu Celcius Ke Rankine : {celtorank}')
```

# 4. Tuliskan suatu program menampilkan massa jenis dengan massa dan volume (dari sebuah balok) inputan user seperti gambar5 4a dan 4b. Rumus: p=m/v Screenshot 2024-09-27 192550.png

```python
# NOMER 4
massaBalok = int(input("Masukan massa balok : "))
panjangBalok = int(input("Masukan panjang baalok : "))
lebarBalok = int(input("Masukan lebar balok : "))
tinggiBalok = int(input("Masukan tinggi balok : "))
volumeBalok = panjangBalok*lebarBalok*tinggiBalok
massaJenis = massaBalok/volumeBalok
print(f"Volume balok : {volumeBalok} m3")
print(f"Massa balok : {massaBalok} kg")
print(f"Massa jenis balok tersebut : {massaJenis} kg/m3")
```

# 5. Tuliskan suatu program yang membaca bilangan integer antara 10 sampai 99 dan kemudian hitunglah hasil pangkat digit 1 yang di pangkatkan dengan digit 2.

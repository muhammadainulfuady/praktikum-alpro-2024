## No1
Tuliskan suatu program yang meminta pengguna untuk memasukkan panjang dari tengah pentagon (segi lima) ke vertex menggunakan fungsi evaal(), dan kemudian hitunglah luas pentagon tersebut, sebagaimana diperlihatkan pada gambar di bawah ini.

Screenshot 2024-10-04 101845.png

```python
import math
r = eval(input("Masukkan panjang dari tengah pentagon ke vertex (r): "))
s = 2 * r * math.sin(math.pi / 5)
luas = 3 * math.sqrt(3) * math.pow(s, 2) / 2
print(f"Luas pentagon tersebut adalah: {luas}")
```

## No 2

Buatlah sebuah program Python yang dapat menghitung panjang sisi miring dari sebuah segitiga siku-siku jika di ketahui panjang sisi depan (opposite) dan panjang sisi samping (adjancent). Program harus menggunakan Teorema Pythagoras  dan fungsi sqrt dari library math untuk menghitung panjang sisi miring.

Screenshot 2024-10-03 223753.png

```python
import math
opposite = eval(input("Masukkan panjang sisi depan (opposite) dalam cm: "))
adjacent = eval(input("Masukkan panjang sisi samping (adjacent) dalam cm: "))
hypotenuse = math.sqrt(math.pow(opposite, 2) + math.pow(adjacent, 2))
print(f"Panjang sisi miring (hypotenuse) adalah: {hypotenuse} cm", )
```

## No 3

Sebuah kerucut memiliki jari-jari alas =4 cm dan panjang sisi miring =5 cm dengan sudut kemiringan 0 = 60. Hitunglah volume kerucut tersebut.

**Rumus**: Volume kerucut dapat dihitung menggunakan rumus:
Screenshot 2024-10-03 224941.png

Dimana tinggi h kerucut dapat dihitung menggunakan sudut kemiringan 0:

Screenshot 2024-10-03 225211.png

**Keterangan:**

Screenshot 2024-10-03 225346.png

```python
import math

jari_jari = 4
sisi_miring = 5
sudut = 60
tinggi = sisi_miring * math.cos(math.radians(sudut))
volume = (1/3) * math.pi * math.pow(jari_jari, 2) * tinggi
print(f"Tinggi kerucut: {tinggi} cm")
print(f"Volume kerucut: {volume} cm")
```

##No 1
Buatkan kode program yang meminta kepada pengguna untuk memasukkan suatu
integer dan periksa apkah bilangan tersebut dapat dibagi 3 dan 7, dapat dibagi 3 atau
7, atau tidak dapat dibagi oleh keduanya!

Contoh tampilan saat program dijalankan:

Screenshot 2024-10-10 211446.png

```python
#Jawaban No1
angka = int(input("Masukkan bilangan = "))
print(f"Apakah {angka} dapat dibagi 3 dan 7? {angka % 3 == 0 and angka % 7 == 0}")
print(f"Apakah {angka} dapat dibagi 3 atau 7? {angka % 3 == 0 or angka % 7 == 0}")
print(f"Apakah {angka} tidak dapat dibagi 3 atau 7? {angka % 3 != 0 and angka % 7 != 0}")
```

##No 2
Buatkan kode program yang memainkan game gunting-batu-kertas yang dimana
Gunting dapat memotong kertas, batu dapat merusak gunting, dan kertas dapat
membungkus batu. Program secara acak membangkitkan bilangan 1/2/3 yang

```python
#Jawaban No2
import random
angka_gbk = int(input("Masukkan pilihanmu. 1 = Gunting/ 2 = Batu/ 3 = Kertas: "))
yangkomputerpilih = random.randint(1, 3)
if angka_gbk == 1:
    pilihan_user = "Gunting"
elif angka_gbk == 2:
    pilihan_user = "Batu"
elif angka_gbk == 3:
    pilihan_user = "Kertas"
if yangkomputerpilih == 1:
    pilihan_komputer = "Gunting"
elif yangkomputerpilih == 2:
    pilihan_komputer = "Batu"
elif yangkomputerpilih == 3:
    pilihan_komputer = "Kertas"
print(f"Komputer memilih {pilihan_komputer} dan Anda memilih {pilihan_user}")
if angka_gbk == yangkomputerpilih:
    print("Hasilnya Anda\nSeri")
elif (angka_gbk == 1 and yangkomputerpilih == 3) or (angka_gbk == 2 and yangkomputerpilih == 1) or (angka_gbk == 3 and yangkomputerpilih == 2):
    print("Hasilnya Anda\nMenang")
else:
    print("Hasilnya Anda\nKalah")
```

##No 3
Suatu persamaan kuadrat dapat mengembalikan dua akar. Pada persamaan ax
2 +bx +
c=0, kedua akar tersebut dapat diperoleh dengan rumus berikut:

Screenshot 2024-10-10 212049.png

b
2 −4a disebut sebagai diskriminan dari persamaan kuadrat. Jika bernilai positif maka
persamaan mempunyai dua akar ril. Jika bernilai nol maka persamaan hanya
mempunyai satu akar. Jika diskriminan ini negatif maka persamaan mempunyai akar tidak ril.

Screenshot 2024-10-10 212220.png

Tuliskan suatu program yang meminta kepada pengguna untuk memasukkan nilai a, b,
dan c dan tampilkan hasilnya berdasarkan pada diskriminan. Jika diskriminannya
positif, tampilkan dua akar. Jika 0 tampilkan satu akar. Jika tidak keduanya, tampilkan
bahwa persamaan tidak mempunyai akar ril

Di bawah ini adalah contoh eksekusi dari program nantinya:

Screenshot 2024-10-10 212409.png

```python
#Jawaban  No3
import math
a, b, c = eval(input("Masukkan a, b, c: "))
discriminant = math.pow(b, 2) - 4 * a * c

if discriminant > 0:
    r1 = (-b + math.sqrt(discriminant)) / (2 * a)
    r2 = (-b - math.sqrt(discriminant)) / (2 * a)
    print(f"Akarnya adalah {r1} dan {r2}")
elif discriminant == 0:
    r1 = -b / (2 * a)
    print(f"Akarnya adalah {r1}")
else:
    print("Persamaan tidak mempunyai akar rill")
```

##No 4
Program akan menerima sebuah bilangan bulat n. Jika n merupakan satuan, cetak string
'satuan'. Jika n merupakan puluhan, cetak string 'puluhan'. Jika n merupakan ratusan,
cetak string 'ratusan'. Jika n merupakan ribuan, cetak string 'ribuan'. Jika n merupakan
puluh ribuan, cetak string 'puluhribuan'.

Screenshot 2024-10-10 212608.png

```python
#Jawaban No4
n = int(input("Masukkan angka: "))
if 0 <= n <= 9:
    print("Satuan")
elif 10 <= n <= 99:
    print("Puluhan")
elif 100 <= n <= 999:
    print("Ratusan")
elif 999 <= n <= 9999:
    print("Ribuan")
elif 9999 <= n <= 99999:
    print("Puluhribuan")
else:
    print("Angka di luar rentang yang diperiksa")
```

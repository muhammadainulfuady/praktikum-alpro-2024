##No 1
Buatlah code yang mampu menampilkan ada berapa banyak suatu huruf dalam satu kalimat seperti di contoh. Keduanya merupakan inputan dari user

Screenshot 2024-10-30 232719.png

```python
#Jawaban No1
kata = input("Masukkan kata atau kalimat: ")
karakter = input("Masukkan karakter: ")

jumlah_karakter = 0

for i in kata:
    if i == karakter:
        jumlah_karakter += 1
print(f'Jumlah karakter {karakter} pada "{kata}" adalah {jumlah_karakter}')
```

##No 2
Buatlah code untuk menambahkan suatu data numerik dalam variabel List, kemudian carilah nilai terbesar dari data list yang sudah ditambahkan!

Screenshot 2024-10-30 232831.png

```python
#Jawaban No2
list_sementara = [2, 4, 6, 8.1, 10]
bilangan_terbesar = 0

print(f"List sementara {list_sementara}")
jumlah_data_tambah = int(input("Masukkan jumlah data yang akan ditambahkan: "))

for i in range(jumlah_data_tambah):
  data_baru = eval(input(f"Data ke-{i + 1} yang ditambahkan : "))
  list_sementara.append(data_baru)

for i in list_sementara:
  if i > bilangan_terbesar:
    bilangan_terbesar = int(i)
print(f"List sekarang {list_sementara}")
print(f"Bilangan terbesar adalah {bilangan_terbesar}")
```

##No 3
Buatlah code untuk menentukan hewan - hewan tersebut apakah hidup di Darat atau di Air, **Simpanlah data hewan dalam dictionary**

Screenshot 2024-10-30 233137.png

```python
#Jawaban No3
hewan = {
    "a" : {"nama" : "Ayam", "lingkungan" : "darat"},
    "b" : {"nama" : "Harimau", "lingkungan" : "darat"},
    "c" : {"nama" : "Hiu", "lingkungan" : "air"},
    "d" : {"nama" : "Salmon", "lingkungan" : "air"}
    }

print(hewan["b"]["lingkungan"])

for i in hewan:
  nama_hewan = hewan[i]["nama"]
  print(f"{i} = {nama_hewan}")

pilihan = input("Pilih Hewanmu : (a/b/c/d) ")
nama_hewan = hewan[pilihan]["nama"]
lingkungan_hewan = hewan[pilihan]["lingkungan"]

print(f"{nama_hewan} Merupakan hewan yang hidup di {lingkungan_hewan}")
```

##No 4
Buatlah code untuk menghitung berapa banyak angka yang memiliki nilai lebih dari rata-rata di dalam suatu data list!

Screenshot 2024-10-30 233458.png

```python
#Jawaban No4
kumpulan_angka = [1, 2, 2, 2.6, 3, 4, 5, 6, 7]
panjang = 0
rata_rata = 0
angka_besar = 0

for i in kumpulan_angka:
  rata_rata += i
  panjang += 1

rata_rata /= panjang

for i in kumpulan_angka:
  if i > rata_rata:
    angka_besar += 1
print(kumpulan_angka)
print(f"Rata - rata dari data tersebut adalah {rata_rata}")
print(f"Jadi banyak angka yang memiliki nilai lebih dari {rata_rata} sebanyak {angka_besar} angka")
```

Buatlah sebuah fungsi panjang yang menerima parameter c. Fungsi ini akan
meminta pengguna memasukkan c jumlah string, kemudian mengurutkan dua string tersebut berdasarkan panjangnya.

Screenshot 2024-11-06 141554.png

```python
def panjang(a,b) :
    if len(a) < len(b) :
        return f"urutan sting dari yang terpendek adalah {a} , {b}"
    elif len(a) > len(b):
        return f"urutan sting dari yang terpendek adalah {b} , {a}"
    else :
        return f"urutan sting dari yang terpendek adalah {a} , {b}"
kondisi = True
while kondisi :
    print("apakah anda ingin membandingkan 2 string[1/2]?")
    pilih = int(input("1.ya\n2.tidak\n=> "))
    if pilih == 1 :
        user1 = input("masukan string yang anda bandingkan : ")
        user2 = input("masukan string yang anda bandingkan  : ")
        print(panjang(user1,user2))
    elif pilih == 2 :
        kondisi = False
    else:
        print("inputan tidak valid program di hentikan")
        kondisi = False
```

##No 2
Buatlah sebuah program Python yang menerima suhu dalam derajat Celsius dari pengguna, lalu menawarkan pilihan untuk mengonversi suhu tersebut ke berbagai skala suhu lainnya. Program ini harus memiliki empat fungsi untuk mengonversi suhu dari Celsius ke skala lainnya, yaitu Fahrenheit, Kelvin, Reamur, dan Rankine.

Contoh:

Screenshot 2024-11-06 141753.png

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def celsius_to_kelvin(celsius):
    return celsius + 273.15

def celsius_to_reamur(celsius):
    return celsius * 4/5

def celsius_to_rankine(celsius):
    return (celsius + 273.15) * 9/5

def konversi_suhu():
    celsius = float(input("Masukkan nilai Celsius: "))
    while True:
        print("\nKonversikan dalam:")
        print("1. Fahrenheit")
        print("2. Kelvin")
        print("3. Reamur")
        print("4. Rankine")
        print("5. Exit")

        pilihan = input("Masukkan pilihan: ")

        if pilihan == "1":
            fahrenheit = celsius_to_fahrenheit(celsius)
            print(f"Konversi dari Celsius ke Fahrenheit adalah {fahrenheit}")
        elif pilihan == "2":
            kelvin = celsius_to_kelvin(celsius)
            print(f"Konversi dari Celsius ke Kelvin adalah {kelvin}")
        elif pilihan == "3":
            reamur = celsius_to_reamur(celsius)
            print(f"Konversi dari Celsius ke Reamur adalah {reamur}")
        elif pilihan == "4":
            rankine = celsius_to_rankine(celsius)
            print(f"Konversi dari Celsius ke Rankine adalah {rankine}")
        elif pilihan == "5":
            print("Terima kasih, program selesai.")
            break
        else:
            print("Pilihan tidak valid, silakan pilih 1 hingga 5.")

konversi_suhu()
```

##No 3
Buatlah sebuah program Python yang memungkinkan pengguna untuk memasukkan
data angka dan string, kemudian mengurutkannya menggunakan fungsi. Program ini harus memiliki dua fungsi yang berbeda: satu untuk mengurutkan angka dan satu lagi untuk mengurutkan string berdasarkan panjangnya.

Contoh:

Screenshot 2024-11-06 142241.png

```python
tampunganAngka = []
tampunganString = []
def string() :
    s = input("masukan string : ")
    tampunganString.append(s)
    return f'string yang anda masukan adalah {s}'

def angka() :
    a = input("masukan angka : ")
    tampunganAngka.append(a)
    return f'angka yang anda masukan adalah {a}'

kondisi = True
while kondisi :
    pilihan = int(input('menu\n1.string\n2.angka\n3.tampilkan yang anda masukan\n4.Exit in program\nmasukan pilihan anda : '))

    if pilihan == 1 :
        print(string())
    elif pilihan == 2 :
        print(angka())
    elif pilihan == 3 :
        print(f'angka yang anda masukan adalah {tampunganAngka}')
        print(f'string yang anda masukan adalah {tampunganString}')
    elif pilihan == 4 :
        kondisi = False
        print("Terimakasih sudah mencoba program saya")
    else :
        print("inputan anda salah program di hentikan")
        kondisi = False
```

##No 4
Buatlah sebuah program Python yang menghitung faktorial dari sebuah bilangan bulat positif menggunakan fungsi rekursif. Faktorial dari sebuah bilangan bulat positif nnn (dilambangkan dengan n!n!n!) didefinisikan sebagai hasil perkalian semua bilangan bulat positif dari 1 hingga n.
Contoh:

Screenshot 2024-11-06 142603.png

```python
def faktorial(a):
    if a == 0:
        return 1
    else :
        return a * faktorial(a-1)
user = int(input("masukan bilangan untuk menghitung faktorial : "))
print(f'faktorial dari {user} adalah {faktorial(user)}')
```
